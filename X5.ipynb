### Вот это значимая фича, сумма заработанных и потраченных баллов

```python
%%time
groupedby_client_data = data.groupby('client_id', as_index=False)
sum_regular_points_received = groupedby_client_data['regular_points_received'].sum() \
    .rename(columns={'regular_points_received': 'sum_regular_points_received'})
sum_express_points_received = groupedby_client_data['express_points_received'].sum() \
    .rename(columns={'express_points_received': 'sum_express_points_received'})
sum_regular_points_spent = groupedby_client_data['regular_points_spent'].sum() \
    .rename(columns={'regular_points_spent': 'sum_regular_points_spent'})
sum_express_points_spent = groupedby_client_data['express_points_spent'].sum() \
    .rename(columns={'express_points_spent': 'sum_express_points_spent'})
```

```python
%%time
mean_regular_points_received = groupedby_client_data['regular_points_received'].mean() \
    .rename(columns={'regular_points_received': 'mean_regular_points_received'})
mean_express_points_received = groupedby_client_data['express_points_received'].mean() \
    .rename(columns={'express_points_received': 'mean_express_points_received'})
mean_regular_points_spent = groupedby_client_data['regular_points_spent'].mean() \
    .rename(columns={'regular_points_spent': 'mean_regular_points_spent'})
mean_express_points_spent = groupedby_client_data['express_points_spent'].mean() \
    .rename(columns={'express_points_spent': 'mean_express_points_spent'})
```
